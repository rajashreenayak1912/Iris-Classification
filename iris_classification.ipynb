```python
import numpy as np
import pandas as pd
import pandas.plotting
import matplotlib.pyplot as plt
import missingno as msno

import seaborn as sns


import warnings
warnings.filterwarnings('ignore')


plt.style.use("fivethirtyeight")
%matplotlib inline
df=pd.read_csv(r'\Users\rajashree\Downloads\Iris.csv')
df.head()
 
```

```python
df.info()  
```

```python
df.drop('Id',axis=1,inplace=True)
df.head()
```

```python
df['Species'].value_counts()
```

```python

sns.countplot(df['Species'].value_counts())
plt.title("species count")
```

```python
msno.bar(df)
```

```python
plt.figure(figsize=(20,5))
sns.boxplot(x='Species',y='SepalLengthCm',data=df.sort_values('SepalLengthCm',ascending=False))
plt.title("The relationship between species and sepallength")
```

```python
plt.figure(figsize=(30,2))
df.plot(kind='scatter',x='SepalWidthCm',y='SepalLengthCm')
plt.title("The Relationship between species and sepalwidth")
```

```python
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=df, size=10)
```

```python
sns.pairplot(df, hue="Species", size=5)
```

```python
df.boxplot(by="Species", figsize=(10, 10))
plt.title("box plot grouped by species")
```

```python
from pandas.plotting import andrews_curves
andrews_curves(df, "Species")
```

```python
plt.figure(figsize=(15,15))
sns.catplot(x='Species',y='SepalWidthCm',data=df.sort_values('SepalWidthCm',ascending=False),kind='boxen',hue="Species")
plt.title("categorical plotting")
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.violinplot(x='Species',y='PetalLengthCm',data=df)
plt.subplot(2,2,2)
sns.violinplot(x='Species',y='PetalWidthCm',data=df)
plt.subplot(2,2,3)
sns.violinplot(x='Species',y='SepalLengthCm',data=df)
plt.subplot(2,2,4)
sns.violinplot(x='Species',y='SepalWidthCm',data=df)
```
